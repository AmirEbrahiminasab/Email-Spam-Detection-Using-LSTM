```python
import pandas as pd
```

```python
data_frame = pd.read_csv("spam_ham_dataset.csv")
data_frame
```

```python
labels = list(data_frame['label_num'])
labels
```

```python
ls = list(data_frame["text"])
ls
```

```python
len(ls)
```

```python
cleaned_ls = [ls[idx].replace('\r\n', ' ') for idx in range(len(ls))]
cleaned_ls
```

```python
from torchtext.data.utils import get_tokenizer
from torchtext.vocab import build_vocab_from_iterator


tokenizer = get_tokenizer("basic_english")


def yield_tokens(data_iter):
    for text in data_iter:
        yield tokenizer(text)


vocab = build_vocab_from_iterator(yield_tokens(cleaned_ls), specials=['<pad>'])
vocab.set_default_index(vocab['<pad>'])
print("Vocabulary:", vocab.get_stoi())
```

```python
len(vocab)
```

```python
max_len = max(len(s) for s in cleaned_ls)
max_len
```

```python
import torch
import torch.nn as nn
```

```python
emails = []

for s in cleaned_ls:
  tokens = tokenizer(s)
  indices = torch.tensor([vocab[token] for token in tokens])

  padded_indices = nn.functional.pad(indices, (0, max_len - len(indices)), value=vocab["<pad>"])

  emails.append(padded_indices)

```

```python
emails
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(emails, labels, test_size=0.4, random_state = 42)
```

```python
email_label_pairs = [(x_train[idx], y_train[idx]) for idx in range(len(x_train))]
```

```python
from torch.utils.data import Dataset, DataLoader
import torch.optim as optim
```

```python
class SpamEmailDataset(Dataset):
    def __init__(self, email_label_pairs, vocab, max_seq_length):
        self.email_label_pairs = email_label_pairs
        self.vocab = vocab
        self.max_seq_length = max_seq_length

    def __len__(self):
        return len(self.email_label_pairs)

    def __getitem__(self, idx):
        email, label = self.email_label_pairs[idx]
        # email_tensor = torch.tensor(email, dtype=torch.long)
        label = torch.tensor(label, dtype=torch.long)

        return email, label



dataset = SpamEmailDataset(email_label_pairs, vocab, max_len)
```

```python
dataloader = DataLoader(dataset, batch_size=16, shuffle=True)
```

## Let's try it without DropOut.

```python
class SpamDetectorModel(nn.Module):
  def __init__(self, embed_size) -> None:
    super(SpamDetectorModel, self).__init__()
    self.embedding = nn.Embedding(len(vocab), embed_size)
    self.lstm = nn.LSTM(embed_size, 64, batch_first=True)
    self.fc = nn.Linear(64, 32)
    self.relu = nn.ReLU()
    self.out = nn.Linear(32, 1)
    self.sigmoid = nn.Sigmoid()


  def forward(self, x):
    embedded = self.embedding(x)
    lstm_out, _ = self.lstm(embedded)
    pooled = torch.max(lstm_out, 1)[0]
    fc = self.fc(pooled)
    out = self.out(fc)

    return self.sigmoid(out)



model = SpamDetectorModel(150)
criterion = nn.BCELoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
model = model.to(device)
```

```python
model.train()
```

```python
num_epoch = 30

for _ in range(num_epoch):
  correct_label, total_label = 0, 0
  for email, label in dataloader:
    email, label = email.to(device), label.to(device)

    output = model(email)

    predicted = (output.squeeze() > 0.5).float()
    total_label += label.size(0)
    correct_label += (predicted == label).sum().item()

    loss = criterion(output.squeeze(), label.float())
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()

  print(f"Train Accuracy on {_+1} epoch is: {100*correct_label / total_label}%.")
  if correct_label / total_label == 1.0:
    break
```

```python
test_email_label_pairs = [(x_test[idx], y_test[idx]) for idx in range(len(x_test))]

test_dataset = SpamEmailDataset(test_email_label_pairs, vocab, max_len)

test_dataloader = DataLoader(test_dataset, batch_size=16, shuffle=True)
```

```python
model.eval()

total_label, correct_label = 0, 0

with torch.no_grad():
  for email, label in test_dataloader:
      email, label = email.to(device), label.to(device)

      output = model(email)
      predicted = (output.squeeze() > 0.5).float()
      total_label += label.size(0)
      correct_label += (predicted == label).sum().item()


print(f"Test Accuracy is: {100 * correct_label / total_label}%")
```
